### **Overfit: -**
Overfit occurs when model memorises training data.  
Here, Training Accuracy > Testing Accuracy.

### **Underfit: -**
Underfit occurs when a model does not learn at all.  
Here, Training Accurary AND Testing Accuracy are BOTH LOW.

### **BestFit: -**
Bestfit is the best model.  
Here, Testing Accurarcy >= Training Accuracy.

## **Splitting the data: -**

```python
X = df.drop("Price", axis = 1)
y = df["Price"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 555)
```

```python
X_train
```

```python
print("Shape of df:", df.shape)
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

## **Building Multi Linear Regression Machine Learning Model: -**

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
col_index = 0
for i in lr.coef_:
    print(df.columns[col_index])
    print("Slope", round(i, 2))
    col_index += 1
    print("*"*30)
    print("y-intercept:", lr.intercept_)
```

We have finished building the model. y_pred variable above has all the predicted values. It predicts using the formula y = mx + c where it takes all the values of x1 and multiplies with the slope(m) + x2 * m2... + xn * mn + c.

## **Finding Error: -**

```python
# Without Library
sse = sum((y_test - y_pred)**2)
print("Sum of Squared Error:", sse)

mse = sse/len(X_test)
print("Mean Squared Error:", mse)

rmse = np.sqrt(mse)
print("Root Mean Squared Error:", rmse)
```

```python
# With Library
from sklearn.metrics import mean_squared_error
error = np.sqrt(mean_squared_error(y_test, y_pred))
print("Error:", round(error, 2))
```

```python
sns.displot(y_test - y_pred);
```

This means most of the time our model does not make any error. But sometimes the predicted value is off by -300000 to 300000.
